### Приведите названия колонок к нижнему регистру и замените пробелы на знак нижнего подчеркивания.

```python
def change_name(name):
    name = name.replace(' ', '_').lower()
    return name

data = data.rename(columns=change_name)
```

```python
data.head()
```

### Пользователи из каких стран совершили наибольшее число успешных бронирований? Укажите топ-5.

```python
check_data = data[data['is_canceled'] == 0]

top_country = check_data.groupby('country', as_index=False).reservation_status.count().sort_values('reservation_status', ascending=False)
top_country.head()
```
